```python
import pandas
import numpy as np
from sklearn.decomposition import PCA

PATH = '/Users/vlasnikita/Documents/vls/py/ml_py_coursera/'

# Загрузите данные close_prices.csv. 
# В этом файле приведены цены акций 30 компаний на закрытии торгов за каждый день периода.
data = pandas.read_csv(PATH + 'lesson_4/close_prices.csv')
del data['date']
# На загруженных данных обучите преобразование PCA с числом компоненты равным 10. 
# Скольких компонент хватит, чтобы объяснить 90% дисперсии?
pca = PCA(n_components=10)
pca.fit(data)
# sum of first 4 elements gives > 0.9
file1 = open(PATH + 'lesson_4/II_1.txt', 'w')
file1.write('4')

# Примените построенное преобразование к исходным данным и возьмите значения первой компоненты.
X = pca.transform(data)
X_first_components = X[:,0]

# Загрузите информацию об индексе Доу-Джонса из файла djia_index.csv. 
# Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса?
djia = pandas.read_csv(PATH + 'lesson_4/djia_index.csv')
del djia['date']
pearson = round(np.corrcoef(X_first_components, np.array(djia)[:,0])[1,0], 2)
file2 = open(PATH + 'lesson_4/II_2.txt', 'w')
file2.write(str(pearson))

# Какая компания имеет наибольший вес в первой компоненте? Укажите ее название с большой буквы.
# max_component_index = np.argmax(pca.explained_variance_ratio_)
max_component_index = 0
max_company_index = np.argmax(np.absolute(pca.components_[max_component_index]))
heaviest_company = list(data)[max_company_index]
file3 = open(PATH + 'lesson_4/II_3.txt', 'w')
file3.write(heaviest_company)
```
